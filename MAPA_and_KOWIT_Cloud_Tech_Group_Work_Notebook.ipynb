## Overview

This notebook will show you how to create and query a table or DataFrame that you uploaded to DBFS. [DBFS](https://docs.databricks.com/user-guide/dbfs-databricks-file-system.html) is a Databricks File System that allows you to store data for querying inside of Databricks. This notebook assumes that you have a file already inside of DBFS that you would like to read from.

This notebook is written in **Python** so the default cell type is Python. However, you can use different languages by using the `%LANGUAGE` syntax. Python, Scala, SQL, and R are all supported.

```
# CORRUPTION File location and type
file_location = "/FileStore/tables/corruption.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
corruption_df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(corruption_df)
```

```
# Create a view or table

corruption_table = "corruption_csv"

corruption_df.createOrReplaceTempView(corruption_table)
```

# Tourism DataFrame

```
# Sort the data by amount of tourists in ascending order
tourism_df = tourism_df.sort_values("tourists_in_millions", ascending=True)

# Select the top 10 countries
top_10_tourism = tourism_df.tail(10)

# Plot the data
plt.figure(figsize=(20, 10))
plt.barh(top_10_tourism["country"], top_10_tourism["tourists_in_millions"])
plt.xlabel("Tourists in Millions")
plt.ylabel("Country")
plt.title("Top 10 Countries by tourists")

# Show the plot
plt.show()
```

Bar plot shows the first 10 most visited countries from the amount of tourists

```
# Import graphs library
import plotly.graph_objects as go
```

```
# Get the overall average
overall_average = tourism_df['tourists_in_millions'].mean()

# Get the average for each country
country_averages = tourism_df.groupby('country')['tourists_in_millions'].mean()

# Create a new dataframe with the country averages and the difference from the overall average
average_df = pd.DataFrame({'country': country_averages.index,
                         'average': country_averages,
                         'difference': country_averages - overall_average})

# Sort the dataframe by the 'average' column in descending order
average_df = average_df.sort_values(by='average', ascending=False)

# Create the bar chart
fig = go.Figure()
fig.add_trace(go.Bar(x=average_df['country'], y=average_df['difference'], name='Difference from overall average'))

# Add a line of the overall average
fig.add_trace(go.Scatter(x=average_df['country'], y=[overall_average for _ in range(len(average_df))],
                         mode='lines', name='Overall Average'))

# Update the layout
fig.update_layout(xaxis_title='country', yaxis_title='Difference from overall average')

# Show the chart
fig.show()
```

There are only 5 countries that have number of tourists more than the average

```
# Sort the dataframe by 'percentage_of_gdp' in descending order
tourism_df = tourism_df.sort_values(by='percentage_of_gdp', ascending=False)

# Select the top 10 rows of the dataframe
top_10_df = tourism_df.head(10)

# Create the horizontal bar chart
plt.barh(top_10_df['country'], top_10_df['percentage_of_gdp'])

# Import percentage formatter
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

# set the x-axis to percentage format
plt.gca().xaxis.set_major_formatter(PercentFormatter())

# Add labels and title
plt.xlabel('Percentage of GDP')
plt.ylabel('Country')
plt.title('Percentage of GDP from Tourism by Country')

# Show the graph
plt.show()
```

Bar plot shows top 10 countries with the highest GDP percentage from tourism
